```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import seaborn as sns

sims=int(1e5)
final=[]
n=40

for i in range(sims):
    Hat=list(range(1,n+1))
    while len(Hat)>1:
        chosen = rnd.sample(Hat,2)
        new_paper = np.abs(chosen[1]-chosen[0])
        Hat.remove(chosen[1])
        Hat.remove(chosen[0])
        Hat.append(new_paper)
    final.append(Hat[0])

Mean=0

for i in range(len(final)):
    Mean = Mean + final[i]
The_Mean = Mean/(10**5)
    
    
fig,ax = plt.subplots()
bins = np.arange(0, n)-0.5
prob = ax.hist(final, bins,density=True, color='purple')
y=np.linspace(1,n+1, sims)
x= 1/The_Mean*np.exp(-(1/The_Mean)*y)
x1= 2/The_Mean*np.exp(-(1/The_Mean)*y)
plt.plot(y,x, label="exponentional distribution")
plt.plot(y,x1, label="2 times exponentional")
plt.legend()
print(The_Mean)
```

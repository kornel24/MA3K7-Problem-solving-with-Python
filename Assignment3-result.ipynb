```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
import seaborn as sns
from multiprocessing import Pool

def simulate_hat_game(n):
    Hat = list(range(1, n + 1))
    while len(Hat) > 1:
        chosen = rnd.sample(Hat, 2)
        new_paper = np.abs(chosen[1] - chosen[0])
        Hat.remove(chosen[1])
        Hat.remove(chosen[0])
        Hat.append(new_paper)
    return Hat[0]

def run_simulations(num_sims, n):
    with Pool() as pool:
        results = pool.map(simulate_hat_game, [n] * num_sims)
    return results

n=2024
sims = int(1e5)
final = run_simulations(sims, n)

The_Mean = sum(final) / sims

fig, (ax1, ax2) = plt.subplots(2, 1)
bins = np.arange(0, n) - 0.5
prob, _, _ = ax1.hist(final, bins, density=True, color='purple')
print(f"The mean value is approximately {The_Mean:.2f}")
```

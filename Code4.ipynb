```python
import numpy as np
import matplotlib.pyplot as plt


#Checking phase
#Check Probabilities of table, and use it to check k
def Probstep(n):
    base=[1,1/2]
    for i in range (3, n + 1):
        new_step= base[-1]/2 + base[-2]/2
        base.append(new_step)
    return base
n=25
print(Probstep(n))
    
```

```python
#Let's plot the oscillation
x=np.linspace(1,25,25)
n=25
base=[1,1/2]
for i in range (3, n + 1):
    new_step= base[-1]/2 + base[-2]/2
    base.append(new_step)
plt.plot(x,base)
plt.grid('on')
plt.ylabel('Probability of landing on step number')
plt.xlabel('Step number')
plt.savefig('step1.png')
```

```python
#Oscillation from 20 to 25
x=np.linspace(20,25,6)
n=25
base=[1,1/2]
for i in range (3, n + 1):
    new_step= base[-1]/2 + base[-2]/2
    base.append(new_step)
newbase=base[-6:]
plt.plot(x,newbase)
plt.grid('on')
plt.ylabel('Probability of landing on step number')
plt.xlabel('Step number')
plt.savefig('step2.png')
```
